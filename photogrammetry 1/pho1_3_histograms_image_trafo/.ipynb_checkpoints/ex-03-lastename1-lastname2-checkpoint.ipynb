````python
# 1. Import modules
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
````

````python
# 2. Load and visualize image
image = img.imread('./images/test.jpg')
image = np.asarray(image, dtype=np.float32)

# Grayscale = 0.299R + 0.587G + 0.114B
gray_image = 0.299 * image[:,:,0] + 0.587 * image[:,:,1] + 0.114 * image[:,:,2]
gray_image = np.asarray(gray_image, dtype=np.uint8)
image = np.asarray(image, dtype=np.uint8)
print("gray image shape: ", gray_image.shape)

# colar image vs grayscale image
fig = plt.figure()

ax1 = fig.add_subplot(1, 2, 1)
ax1.imshow(image)
ax1.set_title('color image')
ax1.axis("off")

ax2 = fig.add_subplot(1, 2, 2)
ax2.imshow(gray_image, cmap='gray')
ax2.set_title('grayscale image')
ax2.axis("off")

plt.show()
````

````python
# Print out the minimum and maximum value of the displayed image as well as its data type
print("gray image data type: ", gray_image.dtype)
print("gray image max value: ", np.max(gray_image), ", min value: ", np.min(gray_image))
````

````python
# 3. Define your function called histogram(image) here:
def histogram(image):
    """ 2D GrayScale image -> histogram.

    Args:
        image (numpy.array, shape = (col, row)): GrayScale image within [0, 255].

    Returns:
        hist(numpy.array, shape = (256, )): Histogram
    """
    row = image.shape[1]
    col = image.shape[0]
    
    hist = np.zeros(256)
    for i in range(row):
        for j in range(col):
            hist[image[j, i]] =  hist[image[j, i]] + 1
    
    return hist
````

````python
# 4. Define your function called cumulativeHistogram(hist) here:
def cumulativeHistogram(hist):
    """ Histogram -> Cumulative Histogram

    Args:
        hist(numpy.array, shape = (256, )): Histogram

    Returns:
        cum_hist(numpy.array, shape = (256, )): Cumulative Histogram
    """
    cum_hist = np.copy(hist)
    for i in range(len(hist)-1):
        cum_hist[i+1] = cum_hist[i] + cum_hist[i+1]
    return cum_hist
````

````python
# 5. Compute the histogram and cumulative Histogram
hist = histogram(gray_image)
cum_hist = cumulativeHistogram(hist)

fig = plt.figure(figsize=(15,6))

ax1 = fig.add_subplot(1, 2, 1)
ax1.bar(np.arange(256), hist)
ax1.set_title('Histogram')

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(cum_hist, '.')
ax2.set_title('Cumulative Histogram')

plt.show()
````

## B. Histogram Equalization *(7 points)*
You have learned the concept for **histogram equalization** in the lecture. It transforms the image such that all intensities are used equally often. 

**Tasks:**
1. Define a **function** ```imageEqualization(image)``` which takes an image as input and returns the transformed image after applying the histogram equalization. *(3 points)*
2. **Compute** the **equalized** grayscale **cube image** and **visualize** it next to the original image. Set the title of each image properly. **Explain** your **observations** in the markdown cell below. *(2 points)*
3. **Compute** the **histogram** and **cumulative histogram** of the equalized image. **Plot** both histograms. In the markdown cell below **describe** the **differences** with respect to the original histograms. *(2 points)*

````python
# 1. Define your function called imageEqualization(image) here:
def imageEqualization(image):
    """ image Equalization.

    Args:
        image (numpy.array, shape = (col, row)): GrayScale image within [0, 255].

    Returns:
         image (numpy.array, shape = (col, row)): GrayScale image within [0, 255].
    """
    row = image.shape[1]
    col = image.shape[0]
    
    # image -> hist
    hist = histogram(image)
    cum_hist = cumulativeHistogram(hist)
    
    # b = f(a) = (1/k) * H(a) + C
    # k = (N - H(0)) / 255
    # C = -H(0) * 255 / (N - H(0))
    H_0 = hist[0]
    N = row * col
    k = (N - H_0) / 255
    C = -H_0 * 255 / (N - H_0)
    
    equalized_image = np.zeros((col, row))
    for i in range(row):
        for j in range(col):
            equalized_image[j, i] = (1/k) * cum_hist[image[j, i]] + C
            
    equalized_image = np.asarray(equalized_image, dtype=np.uint8)
    
    return equalized_image
````

````python
# 2. Compute equalized image
equalized_image = imageEqualization(gray_image)

fig = plt.figure(figsize=(15,6))

ax1 = fig.add_subplot(1, 2, 1)
ax1.imshow(gray_image, cmap='gray')
ax1.set_title('Original')
ax1.axis('off')

ax2 = fig.add_subplot(1, 2, 2)
ax2.imshow(equalized_image, cmap='gray')
ax2.set_title('Equalization Image')
ax2.axis('off')

plt.show()
````

````python
# 3. Compute the histogram and cumulative histogram of the equalized image
hist = histogram(gray_image)
cum_hist = cumulativeHistogram(hist)

eq_hist = histogram(equalized_image)
eq_cum_hist = cumulativeHistogram(eq_hist)

fig = plt.figure(figsize=(15,6))

ax1 = fig.add_subplot(1, 2, 1)
ax1.bar(np.arange(256), hist)
ax1.plot(cum_hist/10, 'k.')
ax1.set_title('Original')

ax2 = fig.add_subplot(1, 2, 2)
ax2.bar(np.arange(256), eq_hist)
ax2.plot(eq_cum_hist/10, 'k.')
ax2.set_title('Equalization Image')

plt.show()
````
